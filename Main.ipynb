```python
import Remake_Dataset as rd 
import XGBoostModel as xgbm
import Logistic_Model as ml
import ModelPerformace as mp
import pandas as pd
```

```python
Dataset = rd.Remake()
```

```python
New_Data = pd.DataFrame({
    "Light Conditions": ["Daylight"],
    "Number of Casualties": [11],
    "Number of Vehicles": [5],
    "Road Surface Conditions": ["Wet or damp"],
    "Road Type": ["Dual carriageway"],
    "Speed limit": [30],
    "Urban or Rural Area": ["Urban"],
    "Weather Conditions": ["Raining no high winds"]
})

XGboost_Model = xgbm.XGboost_Model(Dataset, New_Data)
Logistic_Model = ml.Logistic_Model(Dataset, New_Data)
```

```python
print(mp.Confusion_Matrix_Test(XGboost_Model.Results_Performance))
print(mp.Confusion_Matrix_Test(Logistic_Model.Results_Performance))
```

```python
print(f'XGBoost Model - Accident Severity Prediction: {XGboost_Model.Predict}')
print(f'Logistic Model - Accident Severity Prediction: {Logistic_Model.Predict}')
```
